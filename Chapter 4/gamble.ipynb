## Reinforcement Learning Textbook Exercise 4.9

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.55
state_size = 100

# Initialize state values
states = [0 for i in range(state_size + 1)]
states[state_size] = 1

# initialize with random values
# for i in range(1, state_size):
#     states[i] = np.random.random()

def draw_state_values(states):
    plt.plot([i for i in range(state_size+1)], states)

def simulate(states):
    threshold = 0.1
    # while delta > threshold:
    for k in range(2000):
        for i in range(1, state_size):
            max_r = 0
            # actions
            for j in range(1, min(i + 1, state_size - i + 1)):
                # reward for each action
                r = p * states[i+j] + (1-p) * states[i-j]
                max_r = max(max_r, r)
            states[i] = max_r
        if k in [0, 1, 2, 3, 10, 32, 50]:
            draw_state_values(states)
    draw_state_values(states)
    plt.show()
    return states

def optimal_policy(states):
    policy = []
    for i in range(1, state_size):
        # get best policy
        rewards = dict()
        for j in range(1, min(i+1, state_size-i+1)):
            # rewards for each eaction
            rewards[j] = p * states[i+j] + (1-p) * states[i-j]
        best_move = max(rewards, key=lambda x: rewards[x])
        policy.append(best_move)
    plt.bar([i for i in range(1, state_size)], policy)
    plt.show()

simulate(states)
plt.show()
optimal_policy(states)
```
